```python
import numpy as np
```

Arraay creation and conversion

```python
arr1 = np.array([1,2,3])
print("array:", arr1)
```

```python
print(arr1.size)
```

```python
print(arr1.shape)
```

```python
arr2 = np.array([[1,2],[2,1],[3,4]])
print(arr2.size) #3rows and 2columns
```

```python
arr3 = np.array([[1,2], [2,1], [3,4]])
arr4 = np.array([[5,6], [7,8], [9,0]])
arr34_add = arr3+arr4
print("array addition:", arr34_add)
arr34_mult = arr3*arr4
print("array multiplication:", arr34_mult)
arr5 = arr4%arr3
print("array modulo:",arr5)
```

```python
arr_sum = np.sum([arr3, arr4])
print(arr_sum)
```

```python
b = np.asarray((4,5,6))
print(b)
```

```python
c = np.arange(0, 10, 2)
print(c)
```

```python
d = np.linspace(0,10,3)
print(d)
```

```python
e = np.logspace(0,2,5)
print(e)
```

```python
f = np.zeros((2,3))
print(f)
```

```python
g = np.ones((2, 3), dtype=int)
print(g)
```

```python
h = np.empty((2,2))
print(h)
```

```python
i = np.full((2,2),7)
print(i)
```

```python
j = np.eye(3,4)
print(j)
```

```python
k = np.identity(3)  #identity matrix of order 3
print(k)
```
